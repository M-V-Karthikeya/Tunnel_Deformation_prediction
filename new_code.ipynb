```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

df1 = pd.read_csv('Atal Data (2).xlsx - 2392.tsv',sep='\t')

X1 = []
for i in range(len(df1)):
    X1.append([df1['Time'][i],df1['Go'][i],df1['G1'][i],df1['nk1'][i],df1['Tk1'][i]])
print(X1)
Y1 = []
for i in range(len(df1)):
    Y1.append(df1['Deformation'][i])
print(Y1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X1 = scaler.fit_transform(X1)
```

```python
df2 = pd.read_csv('Atal Data (2).xlsx - 2011.tsv',sep='\t')

X2 = []
for i in range(len(df2)):
    X2.append([df2['Time'][i],df2['Go'][i],df2['G1'][i],df2['nk1'][i],df2['Tk1'][i],df2['Tk2'][i],df2['G2'][i],df2['nk2'][i]])
print(X2)
Y2 = []
for i in range(len(df2)):
    Y2.append(df2['Deformation'][i])
print(Y2)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X2 = scaler.fit_transform(X2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
def lr_p(X,Y,n,df):
    poly = PolynomialFeatures(n)
    x_poly = poly.fit_transform(X)
    model = LinearRegression()
    model.fit(x_poly,Y)
    y_pred = model.predict(x_poly)

    print(list(y_pred))
    print(list(Y))
    plt.scatter(df['Time'],Y)
    plt.scatter(df['Time'],y_pred,alpha=0.4)
    plt.plot()
    return model
```

```python
model2 = lr_p(X1,Y1,31,df1)
```

```python
model = lr_p(X2,Y2,23,df2)
```

```python
def pred(X):
    poly = PolynomialFeatures(23)
    X = poly.fit_transform(X)
    y_pred = model.predict(X)
    return y_pred

df_test = pd.read_csv('test.csv')
X = list(df_test['X'])
y_test = list(df_test['Y'])
X_test = []
for i in X:
    l = []
    l.append(i)
    for i in range(7):
        l.append(0)
    X_test.append(l)
X_test = scaler.fit_transform(X_test)
y_pred = pred(X_test)
plt.scatter(list(df_test['X']),y_test)
plt.scatter(list(df_test['X']),y_pred,alpha=0.4)
plt.title('Linear Regression with Polynomial Features')
plt.show()
```

```python
from keras.layers import Dense
from keras import Sequential
from keras.losses import MeanSquaredError
from keras.activations import linear,relu,softmax,exponential

def nn(X,Y,df):
    layer_1 = Dense(5,exponential)
    a1 = layer_1(np.array(X))
    layer_2 = Dense(3,exponential)
    a2 = layer_2(a1)
    layer_3 = Dense(1,exponential)
    a3 = layer_3(a2)

    model = Sequential([layer_1,layer_2,layer_3])
    model.compile(loss=MeanSquaredError())
    x = np.array(X)
    y = np.array(Y)
    model.fit(x,y,epochs=7500)
    y_pred = model.predict(x)

    rmse = 0
    for i in range(len(y_pred)):
        rmse+=(y[i]-y_pred[i])**2
    rmse = (rmse/len(y_pred))**0.5
    print('error = ',rmse)
    plt.scatter(list(df['Time']),Y)
    plt.scatter(list(df['Time']),y_pred,alpha=0.4)
    plt.show()
    return model
```

```python
model_nn = nn(X2,Y2,df2)
```

```python
y_pred_nn = model_nn(X_test)
plt.scatter(list(df_test['X']),y_test)
plt.scatter(list(df_test['X']),y_pred_nn,alpha=0.4)
plt.title('Neural Network')
plt.show()
```

```python

k = [15]
error = 10**9
y_pred_knn = []
for n in k:
    print(n)
    y_pred = []
    min_dist = []
    for i in X_test:
        dist = []
        for j in range(len(X2)):
            d = 0
            for ind in range(8):
                d+=(i[ind]-X2[j][ind])**2
            d = d**0.5
            dist.append([d,j])
        dist = sorted(dist)
        min_dist.append(dist[:n])
    for i in min_dist:
        yp = 0
        for j in i:
            yp+=Y2[j[1]]
        y_pred.append(yp/n)
    err = 0
    for i in range(len(y_pred)):
        err+=(y_pred[i]-y_test[i])**2
    if(err<error):
        error = err
        y_pred_knn = y_pred
    else:
        continue

plt.scatter(list(df_test['X']),y_test)
plt.scatter(list(df_test['X']),y_pred_knn,alpha=0.4)
plt.title('KNN')
plt.show()
```

```python
y_pred_bag = []
for i in range(len(y_pred)):
    y_pred_bag.append((y_pred[i]+y_pred_knn[i]+y_pred_nn[i])/3)
plt.scatter(list(df_test['X']),y_test)
plt.scatter(list(df_test['X']),y_pred_bag,alpha=0.4)
plt.title('Bagging')
plt.show()
```
